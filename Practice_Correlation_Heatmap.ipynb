```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/RoyMachineLearning/Suicide-Rates-Overview-1985-to-2016/master/master.csv')
```

```python
sns.heatmap(data=pearson,
            xticklabels=pearson.columns,
            yticklabels=pearson.columns, 
            annot=True,
            linewidths=0.8)
```
