```python
mn = df['age'].mean()
sd = df['age'].std()
n = 1
ll = mn-(n)*sd
ul = mn+(n)*sd
df1 = df[(df['age']>ll) & (df['age']<ul)]
print('Total values of age: ',df1['age'].count())
print('Percentage of values in {} std {}'.format(n,(df1['age'].count()/df.shape[0])*100))
```
